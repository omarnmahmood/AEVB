# Visualise "Manifold"

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

def visualize_scatter_with_images(X_2d_data, images, img_name = None, figsize=(20,20), image_zoom=1, labels = None):
    fig, ax = plt.subplots(figsize=figsize)
    artists = []
    #cmap_list = ['Blues_r','Oranges_r','Greens_r','Reds_r','Purples_r','copper','PuRd_r','Greys_r','winter','cool']
    cmap_list = ['viridis','plasma','Wistia','summer','spring','copper','PuRd_r','autumn','winter','cool']
    for c, (xy, i) in enumerate(zip(X_2d_data, images)):
        x0, y0 = xy
        if labels:
            img = OffsetImage(i, zoom=image_zoom, cmap = cmap_list[labels[c]])
        else:
            img = OffsetImage(i, zoom=image_zoom, cmap = 'gray')
        ab = AnnotationBbox(img, (x0, y0), xycoords='data', frameon=False)
        artists.append(ax.add_artist(ab))
    ax.update_datalim(X_2d_data)
    ax.autoscale()
    if img_name:
        plt.savefig('./Images/' + img_name)
    plt.show()
```

```python
test = test.cuda()
```

```python
optimizer = Adam(test.parameters(),lr=1e-3)
```

```python
train_data,_ = VAE_NN.get_data_loaders(b_size=100)
```

```python
t = time.time()
ep = 50
VAE_NN.train(test,optimizer,train_data, VAE_NN.elbo_loss, epochs = ep)
t_e = time.time() - t
print('Seconds for %d epcohs: %d' % (ep,t_e))
```

```python
embeddings_xy = []
imgs = []
labels = []

for i, (t_data,label_v) in enumerate(train_data):
    if i >= 10:
        break
    for data,label in zip(Variable(t_data), Variable(label_v)):
        imgs.append(data.data.cpu().numpy().reshape(28,28))
        labels.append(label.data[0])
        data = data.view(784)
        m = test.encode(data.cuda())[0]
        embedding = m.data.cpu().numpy()
        embeddings_xy.append(embedding)
```

```python
visualize_scatter_with_images(embeddings_xy, imgs, img_name = '2D_latent_c.png', labels=labels)
```

## 20D with t-SNE

```python
train_data,_ = VAE_NN.get_data_loaders(b_size=100)
```

```python
test = VAE_NN.VAE_Net(latent_size = 20)
```

```python
test.load_state_dict(torch.load('../ModelWeights/MNIST_20D_50e'))
```

```python
test.cuda()
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, perplexity=40.0)
```

```python
embeddings_20D = []
imgs = []
labels = []

for i, (t_data,label_v) in enumerate(train_data):
    if i >= 10:
        break
    for data,label in zip(Variable(t_data), Variable(label_v)):
        imgs.append(data.data.cpu().numpy().reshape(28,28))
        labels.append(label.data[0])
        data = data.view(784)
        m = test.encode(data.cuda())[0]
        embedding = m.data.cpu().numpy()
        embeddings_20D.append(embedding)
```

```python
embeddings_array = np.array(embeddings_20D)
```

```python
tSNE_decomp = tsne.fit_transform(embeddings_array)
```

```python
import matplotlib.colors
```

```python
visualize_scatter_with_images(tSNE_decomp, imgs, labels = labels)
```

```python
df = pd.DataFrame(dict(x=[d[0] for d in tSNE_decomp], y=[d[1] for d in tSNE_decomp], color=labels))
```

```python
#plt.figure(figsize=(20,20))
sns.lmplot(data=df,x='x',y='y',hue='color',fit_reg=False, size=7,aspect=1)
#plt.savefig('./Images/tSNE_20D_')
plt.show()
```
